### Max value

```python
# prompt: Generate a code that reads the data frames as input argument
# Then uses two columns named score and label as argument with default value
# after that gets the window size as argument with default value of 50 then calculate the max value of the scores given in the algorithm and propose that as the threshold value for the array item
# if the score value is greater than the threshold (max value of window) then tell it is anomaly (1) otherwise zero
# and then calculate the meterics f1 and MCC values with ground truth labels
# define precision and recall as variables
# return from the function an object that contains f1 , f2 , mcc
# Do not change the default data frame and use : operator for items not magical syntax

from sklearn.metrics import f1_score, matthews_corrcoef, precision_score, recall_score

def thresholding_max_value(df, score_col='score', label_col='label', window_size=50):
    """
    Analyzes anomalies in a DataFrame using a sliding window approach.

    Args:
        df (pd.DataFrame): The input DataFrame.
        score_col (str): The name of the column containing scores. Defaults to 'score'.
        label_col (str): The name of the column containing ground truth labels. Defaults to 'label'.
        window_size (int): The size of the sliding window. Defaults to 50.

    Returns:
        dict: A dictionary containing F1 score, precision, recall, and MCC.
    """
    # Ensure the required columns exist
    if score_col not in df.columns:
        raise ValueError(f"Score column '{score_col}' not found in the DataFrame.")
    if label_col not in df.columns:
        raise ValueError(f"Label column '{label_col}' not found in the DataFrame.")

    predictions = []
    scores = df[score_col].tolist()
    labels = df[label_col].tolist()

    for i in range(len(scores)):
        # Define the window
        start_index = max(0, i - window_size + 1)
        window = scores[start_index : i + 1]

        # Calculate the threshold (max score in the current window)
        threshold = max(window) if window else -float('inf')

        # Make a prediction
        if scores[i] > threshold:
            predictions.append(1)  # Anomaly
        else:
            predictions.append(0)  # Not an anomaly

    # Calculate metrics
    f1 = f1_score(labels, predictions)
    precision = precision_score(labels, predictions, zero_division=0)
    recall = recall_score(labels, predictions)
    mcc = matthews_corrcoef(labels, predictions)

    return {
        'f1': f1,
        'precision': precision,
        'recall': recall,
        'mcc': mcc
    }
```

### α × Max

```python
from sklearn.metrics import f1_score, matthews_corrcoef, precision_score, recall_score

def thresholding_max_value(df, alpha=1 ,score_col='score', label_col='label', window_size=50):
    """
    Analyzes anomalies in a DataFrame using a sliding window approach.

    Args:
        df (pd.DataFrame): The input DataFrame.
        score_col (str): The name of the column containing scores. Defaults to 'score'.
        label_col (str): The name of the column containing ground truth labels. Defaults to 'label'.
        window_size (int): The size of the sliding window. Defaults to 50.

    Returns:
        dict: A dictionary containing F1 score, precision, recall, and MCC.
    """
    # Ensure the required columns exist
    if score_col not in df.columns:
        raise ValueError(f"Score column '{score_col}' not found in the DataFrame.")
    if label_col not in df.columns:
        raise ValueError(f"Label column '{label_col}' not found in the DataFrame.")

    predictions = []
    scores = df[score_col].tolist()
    labels = df[label_col].tolist()

    for i in range(len(scores)):
        # Define the window
        start_index = max(0, i - window_size + 1)
        window = scores[start_index : i + 1]

        # Calculate the threshold (max score in the current window)
        threshold = alpha * (max(window) if window else 0)

        # Make a prediction
        if scores[i] > threshold:
            predictions.append(1)  # Anomaly
        else:
            predictions.append(0)  # Not an anomaly

    # Calculate metrics
    f1 = f1_score(labels, predictions)
    precision = precision_score(labels, predictions, zero_division=0)
    recall = recall_score(labels, predictions)
    mcc = matthews_corrcoef(labels, predictions)

    return {
        'f1': f1,
        'precision': precision,
        'recall': recall,
        'mcc': mcc
    }
```

```python
# prompt: Like above cell define a function that gets alpha as parameter and prints the table

def print_alpha_max_table(alpha):
    """
    Runs the alpha * Max thresholding for a given alpha value and prints the results table.

    Args:
        alpha (float): The alpha value to use for thresholding.
    """
    alpha_darpa_results = thresholding_max_value(DARPA_df, alpha=alpha)
    alpha_iscx_results = thresholding_max_value(ISCX_df, alpha=alpha)
    alpha_ids2018_results = thresholding_max_value(IDS2018_df, alpha=alpha)
    alpha_ddos2019_results = thresholding_max_value(DDOS2019_df, alpha=alpha)

    results_list = []

    results_list.append({
        'Dataset': 'DARPA (Alpha * Max)',
        'F1': alpha_darpa_results['f1'],
        'Precision': alpha_darpa_results['precision'],
        'Recall': alpha_darpa_results['recall'],
        'MCC': alpha_darpa_results['mcc']
    })
    results_list.append({
        'Dataset': 'ISCX (Alpha * Max)',
        'F1': alpha_iscx_results['f1'],
        'Precision': alpha_iscx_results['precision'],
        'Recall': alpha_iscx_results['recall'],
        'MCC': alpha_iscx_results['mcc']
    })
    results_list.append({
        'Dataset': 'IDS2018 (Alpha * Max)',
        'F1': alpha_ids2018_results['f1'],
        'Precision': alpha_ids2018_results['precision'],
        'Recall': alpha_ids2018_results['recall'],
        'MCC': alpha_ids2018_results['mcc']
    })
    results_list.append({
        'Dataset': 'DDOS2019 (Alpha * Max)',
        'F1': alpha_ddos2019_results['f1'],
        'Precision': alpha_ddos2019_results['precision'],
        'Recall': alpha_ddos2019_results['recall'],
        'MCC': alpha_ddos2019_results['mcc']
    })

    results_df = pd.DataFrame(results_list)

    print(f'alpha = {alpha}')
    display(results_df.style.format({
        'F1': '{:.4f}',
        'Precision': '{:.4f}',
        'Recall': '{:.4f}',
        'MCC': '{:.4f}'
    }))

    return results_df
```

### Percentile

```python
# prompt: write the prev section write a function that calculates percentile
# then write a function to draw a table
# Then write a code that iterates through the values 50 , 75, 95,99 percentile instead of max in prev blocks

import numpy as np

def thresholding_percentile(df, percentile_value=95, score_col='score', label_col='label', window_size=50):
    """
    Analyzes anomalies in a DataFrame using a sliding window approach and percentile thresholding.

    Args:
        df (pd.DataFrame): The input DataFrame.
        percentile_value (int): The percentile value to use for thresholding (e.g., 95 for 95th percentile).
                                Defaults to 95.
        score_col (str): The name of the column containing scores. Defaults to 'score'.
        label_col (str): The name of the column containing ground truth labels. Defaults to 'label'.
        window_size (int): The size of the sliding window. Defaults to 50.

    Returns:
        dict: A dictionary containing F1 score, precision, recall, and MCC.
    """
    # Ensure the required columns exist
    if score_col not in df.columns:
        raise ValueError(f"Score column '{score_col}' not found in the DataFrame.")
    if label_col not in df.columns:
        raise ValueError(f"Label column '{label_col}' not found in the DataFrame.")

    predictions = []
    scores = df[score_col].tolist()
    labels = df[label_col].tolist()

    for i in range(len(scores)):
        # Define the window
        start_index = max(0, i - window_size + 1)
        window = scores[start_index : i + 1]

        # Calculate the threshold (percentile of scores in the current window)
        if window:
            threshold = np.percentile(window, percentile_value)
        else:
            threshold = 0 # Should not happen with window_size >= 1

        # Make a prediction
        if scores[i] > threshold:
            predictions.append(1)  # Anomaly
        else:
            predictions.append(0)  # Not an anomaly

    # Calculate metrics
    f1 = f1_score(labels, predictions)
    precision = precision_score(labels, predictions, zero_division=0)
    recall = recall_score(labels, predictions)
    mcc = matthews_corrcoef(labels, predictions)

    return {
        'f1': f1,
        'precision': precision,
        'recall': recall,
        'mcc': mcc
    }
```

```python
def print_percentile_table(percentile_value):
    """
    Runs the percentile thresholding for a given percentile value and prints the results table.

    Args:
        percentile_value (int): The percentile value to use for thresholding.
    """
    percentile_darpa_results = thresholding_percentile(DARPA_df, percentile_value=percentile_value)
    percentile_iscx_results = thresholding_percentile(ISCX_df, percentile_value=percentile_value)
    percentile_ids2018_results = thresholding_percentile(IDS2018_df, percentile_value=percentile_value)
    percentile_ddos2019_results = thresholding_percentile(DDOS2019_df, percentile_value=percentile_value)

    results_list = []

    results_list.append({
        'Dataset': f'DARPA ({percentile_value}th Percentile)',
        'F1': percentile_darpa_results['f1'],
        'Precision': percentile_darpa_results['precision'],
        'Recall': percentile_darpa_results['recall'],
        'MCC': percentile_darpa_results['mcc']
    })
    results_list.append({
        'Dataset': f'ISCX ({percentile_value}th Percentile)',
        'F1': percentile_iscx_results['f1'],
        'Precision': percentile_iscx_results['precision'],
        'Recall': percentile_iscx_results['recall'],
        'MCC': percentile_iscx_results['mcc']
    })
    results_list.append({
        'Dataset': f'IDS2018 ({percentile_value}th Percentile)',
        'F1': percentile_ids2018_results['f1'],
        'Precision': percentile_ids2018_results['precision'],
        'Recall': percentile_ids2018_results['recall'],
        'MCC': percentile_ids2018_results['mcc']
    })
    results_list.append({
        'Dataset': f'DDOS2019 ({percentile_value}th Percentile)',
        'F1': percentile_ddos2019_results['f1'],
        'Precision': percentile_ddos2019_results['precision'],
        'Recall': percentile_ddos2019_results['recall'],
        'MCC': percentile_ddos2019_results['mcc']
    })

    results_df = pd.DataFrame(results_list)

    print(f'Percentile = {percentile_value}')
    display(results_df.style.format({
        'F1': '{:.4f}',
        'Precision': '{:.4f}',
        'Recall': '{:.4f}',
        'MCC': '{:.4f}'
    }))

    return results_df
```
