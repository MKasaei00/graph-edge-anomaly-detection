```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, \
    roc_curve
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./anoedge_g_DARPA_score.csv', header=None, names=['predicted', 'actual'], sep=' ')
```

```python
data
```

```python
from matplotlib import pyplot as plt
data['actual'].plot(kind='hist', bins=20, title='actual')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
data['predicted'].plot(kind='hist', bins=100, title='predicted')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
filtered_data = data[data['predicted'] < 1000]

from matplotlib import pyplot as plt
filtered_data['predicted'].plot(kind='hist', bins=100, title='predicted')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
filtered_data = data[data['predicted'] < 500]

from matplotlib import pyplot as plt
filtered_data['predicted'].plot(kind='hist', bins=100, title='predicted')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
filtered_data = data[data['predicted'] < 200]

from matplotlib import pyplot as plt
filtered_data['predicted'].plot(kind='hist', bins=100, title='predicted')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
filtered_data = data[data['predicted'] < 5]

from matplotlib import pyplot as plt
filtered_data['actual'].plot(kind='hist', bins=100, title='actual')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
filtered_data = data[data['predicted'] > 10000]

from matplotlib import pyplot as plt
filtered_data['actual'].plot(kind='hist', bins=100, title='actual')
plt.gca().spines[['top', 'right',]].set_visible(False)

filtered_data_wrong_cases = len(filtered_data[filtered_data['actual'] == 0])
print('Wrong cases', filtered_data_wrong_cases)
```

```python
# Distribution of 0 class in histogram

filtered_data = data[data['actual'] == 0]

from matplotlib import pyplot as plt
filtered_data['predicted'].plot(kind='hist', bins=100, title='actual')
plt.gca().spines[['top', 'right',]].set_visible(False)

print(f"max value is {max(filtered_data['predicted'])}")
print(f"Number of values greater than 1000 is {len(filtered_data[filtered_data['predicted'] > 10000])}")
```
