## Sum values with score and threshold

Now I want to analyze the relation between sum of values of the matrix with the anomaly score calculated

```python
# prompt: Generate a function that draws a histogram plot for a data frame with score and label get the label and score column names from args with good default value
# the plot is histogram of distribution of anomaly scores

import matplotlib.pyplot as plt

def plot_anomaly_score_histogram(df, score_col='score'):
  """
  Draws a histogram plot for a data frame showing the distribution of anomaly scores.

  Args:
    df (pd.DataFrame): The input DataFrame containing anomaly scores and labels.
    score_col (str): The name of the column containing anomaly scores (default: 'score').
  """
  plt.figure(figsize=(10, 6))
  plt.hist(df[score_col], bins=100)
  plt.title('Distribution of Anomaly Scores')
  plt.xlabel('Anomaly Score')
  plt.ylabel('Density')
  plt.grid(True, linestyle='--', alpha=0.6)
  plt.show()
```

```python


def plot_sum_values_histogram(df, sum_col='sum'):
  """
  Draws a histogram plot for a data frame showing the distribution of sum values.

  Args:
    df (pd.DataFrame): The input DataFrame containing sum values.
    sum_col (str): The name of the column containing sum values (default: 'sum_values').
  """
  plt.figure(figsize=(10, 6))
  plt.hist(df[sum_col], bins=100)
  plt.title('Distribution of Sum Values')
  plt.xlabel('Sum Value')
  plt.ylabel('Density')
  plt.grid(True, linestyle='--', alpha=0.6)
  plt.show()

# Example usage with the concatenated dataframe:
# plot_sum_values_histogram(df_concat)
```

```python
def plot_sum_values_with_timestamp(df, sum_col='sum'):
  """
  Draws a histogram plot for a data frame showing the distribution of sum values with timestamps.

  Args:
    df (pd.DataFrame): The input DataFrame containing sum values and timestamps.
    sum_col (str): The name of the column containing sum values (default: 'sum_values').
  """
  plt.figure(figsize=(10, 6))
  plt.plot(df.index, df[sum_col])
  plt.title('Distribution of Sum Values with Timestamps')
  plt.xlabel('Timestamp')
  plt.ylabel('Sum Value')
```
