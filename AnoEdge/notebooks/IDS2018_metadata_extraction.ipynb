```python
import pandas as pd
```

```python
directory = '/content/graph-edge-anomaly-detection/datasets/IDS2018/'

max_values = pd.read_csv(directory + 'max values.txt', delimiter='\t', header=None)
min_values = pd.read_csv(directory + 'min values.txt', delimiter='\t', header=None)
q1_values = pd.read_csv(directory + 'q1 values.txt', delimiter='\t', header=None)
q2_values = pd.read_csv(directory + 'q2 values.txt', delimiter='\t', header=None)
q3_values = pd.read_csv(directory + 'q3 values.txt', delimiter='\t', header=None)
sum_values = pd.read_csv(directory + 'sum values.txt', delimiter='\t', header=None)
scores_dataframe = pd.read_csv(directory + 'anoedge_g_IDS2018_score.csv', sep=' ', header=None);
scores_dataframe.columns = ['Score', 'Label']
```

```python
# prompt: I have some data frames and I want to merge them in a way that columns are union of all
# and every row is the same as the row with same index in other dataframe

merged_df = pd.concat([scores_dataframe, min_values, q1_values, q2_values, q3_values, max_values, sum_values], axis=1)
merged_df.columns = ['score','label','min','q1','median','q3','max','sum']
merged_df
```
