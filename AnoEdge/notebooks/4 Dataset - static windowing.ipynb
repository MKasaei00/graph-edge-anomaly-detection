## Window staticaly for less compute

### Init and setup static windowing

```python
# prompt: I want to analyze a long time series  value
# and also I want to process very fast
# So I want to window statically  it means I have window w
# So I process a windows and define a threshold and use it for the whole window in the future
# after that update the value when the window finisheds
# I want you to write a code that windows the data with defualt value of 50 as an arg and then call a function with simple implementation of mean value
# So I want to be able to call the main function for df for different databases so here it is an arg with score and label columns as arg like above functions
# Then calculate f1 , mcc measures
# define seperate function to caluclate a threshold for a window and call it when required
# You have to process a window that is before the current point not something in the future

from sklearn.metrics import f1_score, precision_score, recall_score, matthews_corrcoef

def window_staticaly(df, score_col='score', label_col='label', window_size=50):
    """
    Analyzes anomalies in a DataFrame using a static window approach with a mean threshold.

    Args:
        df (pd.DataFrame): The input DataFrame.
        score_col (str): The name of the column containing scores. Defaults to 'score'.
        label_col (str): The name of the column containing ground truth labels. Defaults to 'label'.
        window_size (int): The size of the static window. Defaults to 50.

    Returns:
        dict: A dictionary containing F1 score, precision, recall, and MCC.
    """
    # Ensure the required columns exist
    if score_col not in df.columns:
        raise ValueError(f"Score column '{score_col}' not found in the DataFrame.")
    if label_col not in df.columns:
        raise ValueError(f"Label column '{label_col}' not found in the DataFrame.")

    predictions = []
    scores = df[score_col].tolist()
    labels = df[label_col].tolist()

    current_threshold = 0  # Initialize threshold
    for i in range(len(scores)):
        # Check if we are at the beginning of a new window
        if i % window_size == 0:
            # Define the window for threshold calculation (before the current point)
            start_index = max(0, i - window_size)
            end_index = i
            window_for_threshold = scores[start_index:end_index]
            current_threshold = calculate_threshold_for_window(window_for_threshold)

        # Make a prediction based on the current static threshold
        if scores[i] > current_threshold:
            predictions.append(1)  # Anomaly
        else:
            predictions.append(0)  # Not an anomaly


    # Calculate metrics
    f1 = f1_score(labels, predictions)
    precision = precision_score(labels, predictions, zero_division=0)
    recall = recall_score(labels, predictions)
    mcc = matthews_corrcoef(labels, predictions)

    return {
        'f1': f1,
        'precision': precision,
        'recall': recall,
        'mcc': mcc
    }

```

### Mean value

```python
def calculate_threshold_for_window(window):
    """
    Calculates the mean of the values in a window.

    Args:
        window (list or array): A list or array of numerical values.

    Returns:
        float: The mean of the values in the window, or 0 if the window is empty.
    """
    if not window:
        return 0

    return sum(window) / len(window)
```

```python
# Example usage with your dataframes:
print("\nResults for DARPA dataset (Static Window Mean Thresholding):")
darpa_static_results = window_staticaly(DARPA_df)
print(f"F1 Score: {darpa_static_results['f1']:.04f}")
print(f"MCC Score: {darpa_static_results['mcc']:.04f}")
print(f"Precision: {darpa_static_results['precision']:.04f}")
print(f"Recall: {darpa_static_results['recall']:.04f}")
```

```python
print("\nResults for ISCX dataset (Static Window Mean Thresholding):")
iscx_static_results = window_staticaly(ISCX_df)
print(f"F1 Score: {iscx_static_results['f1']:.04f}")
print(f"MCC Score: {iscx_static_results['mcc']:.04f}")
print(f"Precision: {iscx_static_results['precision']:.04f}")
print(f"Recall: {iscx_static_results['recall']:.04f}")

print("\nResults for IDS 2018 dataset (Static Window Mean Thresholding):")
ids2018_static_results = window_staticaly(IDS2018_df)
print(f"F1 Score: {ids2018_static_results['f1']:.04f}")
print(f"MCC Score: {ids2018_static_results['mcc']:.04f}")
print(f"Precision: {ids2018_static_results['precision']:.04f}")
print(f"Recall: {ids2018_static_results['recall']:.04f}")

print("\nResults for DDOS 2019 dataset (Static Window Mean Thresholding):")
ddos2019_static_results = window_staticaly(DDOS2019_df)
print(f"F1 Score: {ddos2019_static_results['f1']:.04f}")
print(f"MCC Score: {ddos2019_static_results['mcc']:.04f}")
print(f"Precision: {ddos2019_static_results['precision']:.04f}")
print(f"Recall: {ddos2019_static_results['recall']:.04f}")
```

### Percentile

```python
import numpy as np

def calculate_threshold_for_window(window, percentile=95):
    """
    Calculates a percentile of the values in a window.

    Args:
        window (list or array): A list or array of numerical values.
        percentile (float): The percentile to calculate (e.g., 95 for 95th percentile).

    Returns:
        float: The calculated percentile of the values in the window, or 0 if the window is empty.
    """
    if not window:
        return 0

    return np.percentile(window, percentile)
```

### $\alpha \times Max $

```python
# prompt: calculate_threshold_for_window redifine this function to caluclate alpha to define the item

def calculate_threshold_for_window(window, alpha=0.9):
    """
    Calculates a threshold based on alpha multiplied by the maximum value in a window.

    Args:
        window (list or array): A list or array of numerical values.
        alpha (float): The alpha value (multiplier). Should be between 0 and 1.

    Returns:
        float: The calculated threshold, or 0 if the window is empty.
    """
    if not window:
        return 0

    return alpha * np.max(window)
```
